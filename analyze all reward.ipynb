```python
%matplotlib inline

import matplotlib.pyplot as plt
import seaborn as sns

import numpy as np
import pandas as pd
import scipy.stats as stats

import utils

import importlib
importlib.reload(utils)

import os
```

```python
out_dir = './'
```

```python
experiments = ['d75786f7-data-2018-01-25-big-experiment',
                '18a75a2d-data-2018-03-19-big-experiment',
              ]

n_evidence = 4
```

```python

exps = []
nets = []
turns = []
rewards = []
adjusted_rewards = []

for in_dir in experiments:
    
    data,fails,parts = utils.get_data(in_dir, n_evidence)
    
    num_turns = max(data['turn'])

    for net in sorted(set(data['net'])):
    
        for t in range(1,num_turns+1): 
               
            exps += [in_dir]
            nets += [str(net) + '-' + in_dir]
            turns += [t]
            rewards += [utils.get_reward(data,parts,net,t)]
            adjusted_rewards += [utils.get_reward(data,parts,net,t) - utils.get_reward(data,parts,net,1)]    
            #adjusted_rewards += [utils.get_reward(data,parts,net,t) - np.mean(parts[net] - 0.5)]    
    
df = pd.DataFrame({'experiment':exps,'net':nets,'turn':turns,'reward':rewards,'adjusted':adjusted_rewards})    
```

```python
df
```

```python
sns.set(context = 'paper', font_scale = 3, font='serif', style = 'white',  rc={"lines.linewidth": 2.5})
sns.pointplot('turn','adjusted',data=df)
plt.xlabel('Turn')
plt.ylabel('Adjusted Score')
plt.savefig('score-over-time.jpg', bbox_inches = 'tight')
```
